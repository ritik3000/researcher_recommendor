```python
import pandas as pd
import pickle
```

```python
df=pd.read_json("Main_dataset.json")
```

```python
df.head()
```

```python
df=df.reset_index(drop=True)
```

```python
df.drop(['id'],axis=1,inplace=True)
```

```python
def handle_abstract(old_abstract):
    if 'Background#R##N#' in old_abstract:
        return(old_abstract[16:])
    return(old_abstract)
```

```python
df['abstract'] = df['abstract'].apply(handle_abstract)
df=df.sort_values('year')
df=df.reset_index(drop=True)
```

```python
df
```

```python
df['author'], labels = pd.factorize(df['author'])
```

```python
publication_count=df['author'].value_counts()
```

```python
count=0
author_list_collab_5=[]
for index,val in publication_count.iteritems():
    if(val>5):
        author_list_collab_5.append(tuple((index,val)))
```

```python
len(author_list_collab_5)
```

```python
pickle.dump(author_list_collab_5,open("author_list_collab_5.pkl",'wb'))
```

```python
df.to_json("Main_dataset.json")
```

```python
pickle.dump(labels,open("authors_label_main_dataset.pkl","wb"))
```
