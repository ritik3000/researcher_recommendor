```python
import pandas as pd
import pickle
import time
```

```python
df1=pd.read_json('dblp_0_train.json')
df2=pd.read_json('dblp_1_train.json')
df3=pd.read_json('dblp_2_train.json')
df4=pd.read_json('dblp_3_train.json')
```

```python
frames=[df1,df2,df3,df4]
df=pd.concat(frames)
df.head()
```

```python
with open('Aut_vec.pkl', 'rb') as f:
    u = pickle._Unpickler(f)
    u.encoding = 'latin1'
    p = u.load()
```

```python
data_dict={'Authors':[],'Venue':[],'references':[],'title':[],'abstract':[],'year':[]}
author_list=[]
for i in p:
    author_list.append(i)
```

```python
%%time
flag=1
for ind,i in enumerate(author_list):
    if ind%100==0:
        print(ind%100,end=' ')
    count = 0
    for j in df.iloc[:,1]:
        #print(j)
        if i in j:
            flag=1
            for auth in j:
                if auth not in author_list:
                    flag=0
                    break
            #print(count)
            if flag:
                data_dict['Authors'].append(df.iloc[count, 1])
                data_dict['Venue'].append(df.iloc[count, 5])
                data_dict['references'].append(df.iloc[count, 3])
                data_dict['title'].append(df.iloc[count, 4])
                data_dict['abstract'].append(df.iloc[count, 0])
                data_dict['year'].append(df.iloc[count, 6])
        count +=1
```

```python
df_final=pd.DataFrame(data_dict)
```
