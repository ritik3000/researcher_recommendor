```python
import pandas as pd
import pickle
import json
import math
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_json("jsons/dblp_merge.json")
df.reset_index(inplace=True)
df
```

```python
df['venue'], labels = pd.factorize(df['venue'])
df=df[['authors','venue','year']]
df
```

```python
aut_list=pickle.load(open("pickles/aut_list.pkl","rb"))
```

```python
total_venue=df['venue'].max()
```

```python
alist=[0]*(total_venue+1)
paper_dict={key: list(alist) for key in aut_list}
```

```python
year_max= df['year'].max()
```

```python
coauthor_list=pickle.load(open("pickles/coauthor_list.pickle","rb"))
```

```python
count=0
for i in range(df.shape[0]):
    if count%100000==0:
        print(count//100000,end=' ')
    count+=1
    for author in df.iloc[i].authors:
        try:
            #print(1/math.log(year_max-df.iloc[i].year+2,2))
            paper_dict[author][df.iloc[i].venue]+=(1/math.log(year_max-df.iloc[i].year+2,2))
        except:
            print("",end='')
```

```python
json.dump(paper_dict,open("jsons/log_venue_vector_unnormalised.json","w"))
```

```python
count=0
for author in paper_dict:
    if count%1000==0:
        print(count//1000,end='')
    count+=1
    raw = paper_dict[author]
    s = sum(raw)
    norm = [float(i)/s for i in raw]
    #print(norm)
    paper_dict[author]=norm
```

```python
log_venue_vector_score={key: None for key in coauthor_list}
```

```python
for coauthor in coauthor_list:
    log_venue_vector_score[coauthor]=cosine_similarity(np.array([paper_dict[coauthor[0]]]),np.array([paper_dict[coauthor[1]]]))[0][0]
```

```python
pickle.dump(log_venue_vector_score,open("pickles/log_venue_vector_score.pkl","wb"))
```
