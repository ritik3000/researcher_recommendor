```python
def load_doc(filename):
    # open the file as read only
    file =open(filename,encoding="utf8")
    config=json.loads(file.read().strip())
    if config['text'].find('PM')!=-1:
        config['text']=config['text'][config['text'].find('PM')+2:]
    
    if config['text'].find('AM')!=-1:
        config['text']=config['text'][config['text'].find('AM')+2:]
    # read all text
    # close the file
    file.close()
    return config['text']
```

```python
def load_stories(directory):
    articles = []
    for name in listdir(directory):
        filename = directory + '/' + name
        # load document
        doc = load_doc(filename)
        articles.append(doc)
    return articles
```

```python
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
```

```python
def clean_lines(lines):
    cleaned = list()
    # prepare a translation table to remove punctuation
    table = str.maketrans('', '', string.punctuation)
    for line in lines:
        # strip source cnn office if it exists
        # tokenize on white space
        line = line.split()
        # convert to lower case
        line = [word.lower() for word in line]
        # remove punctuation from each token
        line = [i for i in line if i not in stop]
        line = [ch for ch in line if ch not in exclude]
        line = [lemma.lemmatize(word) for word in line]
        line = [w.translate(table) for w in line]
        # remove tokens with numbers in them
        line = [word for word in line if word.isalpha()]
        # store as string
        cleaned.append(' '.join(line))
    # remove empty strings
    cleaned = [c for c in cleaned if len(c) > 0]
    return cleaned
```

```python
articles=load_stories(r"C:/Users/ritik/Downloads/Compressed/us-financial-news-articles")
print('Loaded articles %d' % len(articles))
```

```python
final=[]
for j in articles:
    rd1=clean_lines(j.split('.'))
    for i in rd1:
        p=i.split()
        final.append(p)  
```
