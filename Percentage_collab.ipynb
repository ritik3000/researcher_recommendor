```python
import pandas as pd
import pickle
import json
import math
```

```python
coauthor_list=pickle.load(open("pickles/coauthor_list.pickle","rb"))
aut_list=pickle.load(open("pickles/aut_list.pkl","rb"))
```

```python
df=pd.read_json("jsons/dblp_merge.json")
df.reset_index(inplace=True)
df
```

```python
alist=[]
paper_dict={key: list(alist) for key in aut_list}
```

```python
author_list=list(df['authors'])
```

```python
count=0
for i in author_list:
    if(count%100000==0):
        print(count//100000,end='')
    count+=1
    for author in i:
        try:
            paper_dict[author].append(list(i))
        except:
            print('',end='')
```

```python
pickle.dump(paper_dict,open("pickles/author_all_collab.pkl","wb"))
```

```python
percentage_collab_score={key: None for key in coauthor_list}
```

```python
count=0
for coauthor in percentage_collab_score:
    if count%10000==0:
        print(count//10000,end=' ')
    count+=1
    tot=len(paper_dict[coauthor[0]])
    rel=0
   # print(tot)
    for authors in paper_dict[coauthor[0]]:
        for author in authors:
            if author==coauthor[1]:
                rel+=1
                break
   # print(rel)
    percentage_collab_score[coauthor]=rel/(2*tot)
    tot=len(paper_dict[coauthor[1]])
    rel=0
    for authors in paper_dict[coauthor[1]]:
        for author in authors:
            if author==coauthor[0]:
                rel+=1
                break
    percentage_collab_score[coauthor]+=rel/(2*tot)
```
