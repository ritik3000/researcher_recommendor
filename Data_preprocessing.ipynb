```python
import pandas as pd
import re
import time
import pickle
```

```python
df1=pd.read_json("jsons/dblp-ref-0.json",lines=True)
df2=pd.read_json("jsons/dblp-ref-1.json",lines=True)
df3=pd.read_json("jsons/dblp-ref-2.json",lines=True)
df4=pd.read_json("jsons/dblp-ref-3.json",lines=True)
frames=[df1,df2,df3,df4]
df=pd.concat(frames)
```

```python
df.head()
```

```python
print("With Nan: ",df.shape)
df.dropna(inplace=True)
print("Without Nan: ",df.shape)
```

```python
df.set_index('id',inplace=True)
```

```python
df
```

```python
df=df.drop(df[df.references.map(len)<3].index)
```

```python
df.shape
```

```python
df=df.drop(df[df.venue.map(len)<1].index)
```

```python
df.shape
```

```python
df
```

```python
s = df.iloc[:, 1]
```

```python
s.shape
new_s=[]
```

```python
for i in s:
    names=re.findall(r'(\'[A-Z a-z\.\-]+\'[, ]*?)+',str(i))
    for name in names:
        new_s.append(name)
        
s1=pd.Series(new_s)
df.head()
```

```python
str(df.iloc[0].authors)
print(s1.shape)
print(s1[:5])
```

```python
df_new={'abstract':[],'author':[],'id':[],'references':[],'title':[],'venue':[],'year':[]}
count=0
index_list=[]
for i in range(s1.shape[0]):
    index_list.append(i+1)
#start_time=time.time()
for i in s1:
    if(count%100000==0):
        print(count)
       # print(time.time()-start_time)
    if str(df.iloc[count].authors).find(i) !=-1:
        #print(i)
        #print(df.iloc[count].year)
        df_new['abstract'].append(df.iloc[count].abstract)
        df_new['author'].append(i)
        df_new['id'].append(df.iloc[count].id)
        df_new['references'].append(df.iloc[count].references)
        df_new['title'].append(df.iloc[count].title)
        df_new['venue'].append(df.iloc[count].venue)
        df_new['year'].append(df.iloc[count].year)
    
    else:
        count+=1
        df_new['abstract'].append(df.iloc[count].abstract)
        df_new['author'].append(i)
        df_new['id'].append(df.iloc[count].id)
        df_new['references'].append(df.iloc[count].references)
        df_new['title'].append(df.iloc[count].title)
        df_new['venue'].append(df.iloc[count].venue)
        df_new['year'].append(df.iloc[count].year)
#print(df_new)
#print(count)
df1=pd.DataFrame(df_new, index=index_list)
df1
```

```python
df1.to_json("Major_dataset_ref.json")
```
