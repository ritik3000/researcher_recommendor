```python
def thread_names(l,r):
    count=0
    for i in range(l,r):
        if count%1000==0:
            print(l," ",r," ",count//1000)
        count+=1
        author_list=df.iloc[i].authors
        author_permutation=[[author_list[i],author_list[j]] for i in range(len(author_list)) for j in range(i+1, len(author_list))]
        for j in author_permutation:
            try:
                idx=coauthor_list.index(j)
                #print(idx)
                venues_score[idx]+=float(float(frequencies[df.iloc[i].venue]-fmin)/float(fmax-fmin))
                #print("venue_score[%d] : %f"%(idx, venues_score[idx]))
            except:
                t=0
```

```python
def multiprocessed():
    processes = []
    n = 0
    threads = 56
    for i in range(0, threads):
        stop = n + math.floor(2408010/threads + 1) if n + threads <= 2408010 else 2408010
        p = Process(target=thread_names, args=(n, stop))
        n = stop + 1
        processes.append(p)
    # Start the processes
    for p in processes:
        p.start()
    # Ensure all processes have finished execution
    count = 0
    for p in processes:
        p.join()
        print("Process %d is over and time taked is : %f" %(count, (time.time() - start)))
        count += 1
```

```python
%%time
start = time.time()
if __name__=="__main__":
    manager = multiprocessing.Manager()
    venues_score = manager.list([0.0 for i in range(375303)])
    multiprocessed()
```
